```python
import pandas as pd
import numpy as np
import os
import xlrd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
import os
%matplotlib inline
```

```python
file=pd.ExcelFile('Case_study_data.xlsx')
```

```python
churn=file.parse('Churn Customers')
active=file.parse('Active Customers ')
```

```python
df = pd.concat([active, churn], ignore_index=True)
```

```python
df
```

```python
df.columns
```

```python
df=df.drop(['SALES_CHANNEL','MARITAL_STATUS'],axis=1)
```

```python
df=df.drop(['AGE','FIRST_NAME','FIRST_COMMUNICATION_DATE','UPGRADE_DATE','LAST_NAME','COMPANY_SOURCE_NAME','DATE_OF_BIRTH','CHURN_DATE','ZIPCODE','UPGRADE_DATE','EMAIL_ADDRESS','ADDRESS','CHURN_DATE','CITY'],axis=1)
```

```python
df
```

```python
df = pd.get_dummies(df,columns=['ACCOUNT_STATUS','SERVICE_PLAN','MODEL', 'DIRECT_INDIRECT_CHANNEL', 'SALES_SOURCE',
       'STATE', 'FEEDBACK', 'SATELLITE', 'SITE_TYPE_DESC',
       'WARRANTY_NAME', 'MOST_RECENT_SALES_CHANNEL', 'GENDER', 'INCOME',
       'PRESENCE_OF_CHILDREN', 'COMPUTER_OWNER'])
```

```python
df.head(20)
```

```python
x=df.loc[:, ['SERVICE_PLAN_Elite','MODEL_HN9000','DIRECT_INDIRECT_CHANNEL_Indirect','STATE_ CA','FEEDBACK_BAD','SATELLITE_SPACEWAY3','SITE_TYPE_DESC_Spaceway','WARRANTY_NAME_0','GENDER_1.0','INCOME_0','PRESENCE_OF_CHILDREN_0.0','COMPUTER_OWNER_Y']]
```

```python
y=df.loc[:, 'ACCOUNT_STATUS_Closed'] 
```

```python
x.isnull().sum()
```

```python
x.describe()
```

```python
x=x.fillna(x.mean())
```

```python
print (x.shape, y.shape)
```

```python
#import model specific libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3, random_state=100)
```

```python
#validate the shape of train and test dataset
print (X_train.shape)
print (y_train.shape)

print (X_test.shape)
print (y_test.shape)
```

```python
logreg =LogisticRegression(class_weight='balanced')
```

```python
sd=logreg.fit(X_train,y_train)
```

```python
logreg.score(X_train,y_train)
```

```python
predictions=logreg.predict(X_test)
```
